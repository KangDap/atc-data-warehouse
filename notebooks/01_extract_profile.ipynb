# Extract & Data Profiling

This notebook reads all raw CSV files from `data/raw`, displays the initial data contents, and creates a basic profiling summary before the transformation stage.

```python
import pandas as pd
from pathlib import Path

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', 120)
```

## Path Configuration

```python
PROJECT_ROOT = Path.cwd() if (Path.cwd() / 'data' / 'raw').exists() else Path.cwd().parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
VALIDATION_DIR = PROJECT_ROOT / 'data' / 'validation'

VALIDATION_DIR.mkdir(parents=True, exist_ok=True)

print('Raw data folder:', RAW_DIR.resolve())
print('Validation folder:', VALIDATION_DIR.resolve())
```

## Raw CSV File Inventory

```python
csv_files = sorted(RAW_DIR.glob('*.csv'))

file_inventory = pd.DataFrame({
    'file_name': [file.name for file in csv_files],
    'table_name': [file.stem for file in csv_files],
    'size_kb': [round(file.stat().st_size / 1024, 2) for file in csv_files],
})

display(file_inventory)
```

## Extract All CSV Files

```python
tables = {}

for file in csv_files:
    table_name = file.stem
    tables[table_name] = pd.read_csv(file)

print(f'Total tables loaded: {len(tables)}')
print(list(tables.keys()))
```

## CSV Content Preview

This section displays the first few rows from each CSV file.

```python
for table_name, df in tables.items():
    print(f'\n=== {table_name} ===')
    print(f'Rows: {df.shape[0]:,}')
    print(f'Columns: {df.shape[1]:,}')
    display(df.head(10))
```

## Initial Column Structure and Data Types

```python
for table_name, df in tables.items():
    print(f'\n=== {table_name} ===')
    schema_df = pd.DataFrame({
        'column_name': df.columns,
        'dtype': [str(dtype) for dtype in df.dtypes],
        'non_null_count': df.notna().sum().values,
        'null_count': df.isna().sum().values,
        'null_percentage': (df.isna().mean().values * 100).round(2),
        'unique_count': df.nunique(dropna=True).values,
    })
    display(schema_df)
```

## Missing Value Check

```python
for table_name, df in tables.items():
    missing_df = (
        df.isna()
        .sum()
        .reset_index()
        .rename(columns={'index': 'column_name', 0: 'missing_count'})
    )
    missing_df['missing_percentage'] = (missing_df['missing_count'] / len(df) * 100).round(2)
    missing_df = missing_df[missing_df['missing_count'] > 0]

    print(f'\n=== {table_name} ===')
    if missing_df.empty:
        print('No missing values found.')
    else:
        display(missing_df.sort_values('missing_count', ascending=False))
```

## Duplicate Row Check

```python
duplicate_summary = []

for table_name, df in tables.items():
    duplicate_rows = int(df.duplicated().sum())
    duplicate_summary.append({
        'table_name': table_name,
        'row_count': len(df),
        'duplicate_rows': duplicate_rows,
        'duplicate_percentage': round(duplicate_rows / len(df) * 100, 2) if len(df) else 0,
    })

duplicate_summary_df = pd.DataFrame(duplicate_summary)
display(duplicate_summary_df)
```

## Extract Profiling Summary

```python
profile_summary = []

for table_name, df in tables.items():
    profile_summary.append({
        'table_name': table_name,
        'row_count': len(df),
        'column_count': len(df.columns),
        'total_missing_values': int(df.isna().sum().sum()),
        'duplicate_rows': int(df.duplicated().sum()),
        'columns': ', '.join(df.columns),
    })

profile_summary_df = pd.DataFrame(profile_summary)
display(profile_summary_df)

output_path = VALIDATION_DIR / 'extract_profile_summary.csv'
profile_summary_df.to_csv(output_path, index=False)

print(f'Profiling summary saved to: {output_path}')
```
